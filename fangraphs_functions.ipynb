```python
import pandas as pd 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import datetime
import sqlite3 
import pybaseball as pyb #Pybaseball is where we will be pulling data from
```

# Fangraphs Hitter Data

```python
fg_hitter = pyb.batting_stats(2023)
```

# Fangraphs Pitcher Data

```python
fg_pitcher = pyb.pitching_stats(2023)
```

### Leaderboards for Best/Worst difference of wOBA vs xwOBA

```python
# Get a small subset of columns to view in the leaderboard
fg_hitter_filt = fg_hitter[['Season','Name','AB','PA','BB%','K%','BB/K','LD%','FB%','HR/FB','CSW%','O-Swing%','Z-Swing%','Z-Contact%','Pull%','Cent%','Oppo%','xwOBA','wOBA']]
fg_hitter_filt['wDIF'] = fg_hitter_filt['wOBA'] - fg_hitter_filt['xwOBA'] 
```

#### Positive difference (outperforming xwOBA)

```python
from IPython.display import display

selected_rows_hitter = fg_hitter_filt[fg_hitter_filt['PA'] > 399]
positive_leaderboard = selected_rows_hitter.sort_values('wDIF', ascending=False).head(20)

# Define the columns to apply the background gradient to
columns_to_color = ['wDIF', 'O-Swing%', 'Z-Swing%','Z-Contact%']  # Add more columns as needed

# Apply background gradient to specific columns in the leaderboard DataFrame
positive_leaderboard = positive_leaderboard.style.background_gradient(cmap='Reds', subset=columns_to_color).format(precision=3)

# Display the styled leaderboard in Jupyter Notebook
display(positive_leaderboard) 
```

#### Negative difference (underperforming xwOBA)

```python
from IPython.display import display

selected_rows_hitter = fg_hitter_filt[fg_hitter_filt['PA'] > 399]
negative_leaderboard = selected_rows_hitter.sort_values('wDIF', ascending=True).head(20)

# Define the columns to apply the background gradient to
columns_to_color = ['wDIF', 'O-Swing%', 'Z-Swing%','Z-Contact%']  # Add more columns as needed

# Apply background gradient to specific columns in the leaderboard DataFrame
negative_leaderboard = negative_leaderboard.style.background_gradient(cmap='Blues_r', subset=columns_to_color).format(precision=3)

# Display the styled leaderboard in Jupyter Notebook
display(negative_leaderboard) 
```
